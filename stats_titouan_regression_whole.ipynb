```python
########################################
# Import Data
########################################
import pandas as pd
import os

#define file to import
file = 'data_regression_dataframe.csv'

#store name for future use
fname, _ = os.path.splitext(file)

#Import Data
df = pd.read_csv(file, sep=',', header=0)

#clean dataframes
df = df.rename(columns=lambda x: x.strip())
df = df.dropna()

#Show Dataframe
print(fname + ' dataset:')
df
```

```python
#####################################
# Separate linear regression
#####################################
from colorama import Fore, Back, Style
import matplotlib .pyplot as plt
plt.rcParams['figure.figsize'] = [12, 7]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
import numpy as np
from scipy.stats import shapiro, pearsonr, linregress

"""
Separate linear regression can now be performed with transformed data using
the only SWE variable that was normally distributed "SWE_raw" (absolute diff between PRE & POST tests)
as dependant variable and all other (including LRELATIVE which is not normal) as independent variables.
"""

#Get a working dataframe
reg_df = df.copy()

#Filter if necessary
#reg_df = reg_df[reg_df.MUSCLE.str.contains('ST|BF')]

#Add some variable dor visualization purpose
reg_df['Marker'] = reg_df['MUSCLE'].replace('SM', 's').replace('ST', 'o').replace('BF', '^')
reg_df['Color'] = reg_df['GROUPE'].replace('NHE', 'red').replace('DL', 'blue')

#define independent variable
dep = 'SWE_raw'
#get colnames of dependent variables
cols = list(reg_df.drop(['ID', 'GROUPE', 'MUSCLE', 'Marker', 'Color', dep], axis=1).columns)
#loop them
for c in cols:
    #create a figure at each column
    fig, axs = plt.subplots(2, 2) #all, group, muscle, interact
    fig.suptitle(c)
    #proceed for all data
    x = np.array(reg_df[dep])
    y = np.array(reg_df[c])
    #all in one for linear regression
    result = linregress(x, y)
    #show results
    print(f'{c:>18}', end="")
    if result.pvalue < 0.05:
        print(Fore.GREEN + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
    elif result.pvalue < 0.1:
        print(Fore.YELLOW + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
    else:
        print(Fore.RED + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
    print(Style.RESET_ALL)
    #draw
    line = f'r={result.rvalue:.2f}, p={result.pvalue:.3f}, N={len(x)}'
    axs[0, 0].scatter(x, y, marker='o', c='#000')
    axs[0, 0].plot(x, result.intercept + result.slope * x, label=line)
    axs[0, 0].set_title('Pooled')
    axs[0, 0].set_ylabel(c)
    axs[0, 0].set_xlabel('SWE raw')
    axs[0, 0].legend(facecolor='white', loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=1)
    for i, txt in enumerate(reg_df['ID']):
        axs[0, 0].annotate(chr(int(txt) + 97), (x[i]+1, y[i]))
    #Then proceed for sub groups (level 2)
    for gr, sub in reg_df.groupby('Color'):
        x = np.array(sub[dep])
        y = np.array(sub[c])
        result = linregress(x, y)
        #show results
        print(f'{c:>18} {sub.GROUPE.iloc[0]}', end="")
        if result.pvalue < 0.05:
            print(Fore.GREEN + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
        elif result.pvalue < 0.1:
            print(Fore.YELLOW + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
        else:
            print(Fore.RED + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
        print(Style.RESET_ALL)
        #draw
        line = f'{sub.GROUPE.iloc[0]}_ r={result.rvalue:.2f}, p={result.pvalue:.3f}, N={len(x)}'
        axs[0, 1].scatter(x, y, marker='o', c=gr)
        axs[0, 1].plot(x, result.intercept + result.slope * x, color=gr, label=line)
        for i, txt in enumerate(sub['ID']):
            axs[0, 1].annotate(chr(int(txt) + 97), (x[i]+1, y[i]))
        #interaction group & muscles (level 3)
        for mu, sub2 in sub.groupby('Marker'):
            x = np.array(sub2[dep])
            y = np.array(sub2[c])
            result = linregress(x, y)
            #show results
            print(f'{c:>18}: {sub2.GROUPE.iloc[0]}x{sub2.MUSCLE.iloc[0]}', end="")
            if result.pvalue < 0.05:
                print(Fore.GREEN + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
            elif result.pvalue < 0.1:
                print(Fore.YELLOW + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
            else:
                print(Fore.RED + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
            print(Style.RESET_ALL)
            #define a color based on marker
            if mu == 's': color = 'green'
            elif mu == 'o': color = 'orange'
            elif mu == '^': color = 'cyan'
            if gr == 'red':
                alpha=1
            elif  gr == 'blue':
                alpha= .3
            #draw
            line = f'{sub2.GROUPE.iloc[0]}x{sub2.MUSCLE.iloc[0]}_ r={result.rvalue:.2f}, p={result.pvalue:.3f}, N={len(x)}'
            axs[1, 1].scatter(x, y, marker=mu, c=color, alpha=alpha)
            axs[1, 1].plot(x, result.intercept + result.slope * x, color=color, alpha=alpha, label=line)
            for i, txt in enumerate(sub2['ID']):
                axs[1, 1].annotate(chr(int(txt) + 97), (x[i]+1, y[i]))
        axs[0, 1].set_title('GROUP')
        axs[0, 1].set_ylabel(c)
        axs[0, 1].set_xlabel(dep)
        axs[0, 1].legend(facecolor='white', loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=1)
        axs[1, 1].set_title('GROUP x MUSCLE')
        axs[1, 1].set_ylabel(c)
        axs[1, 1].set_xlabel(dep)
        axs[1, 1].legend(facecolor='white', loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=2)
    #now for sub muscles (level 2)
    for mu, sub in reg_df.groupby('Marker'):
        x = np.array(sub[dep])
        y = np.array(sub[c])
        result = linregress(x, y)
        #show results
        print(f'{c:>18} {sub.MUSCLE.iloc[0]}', end="")
        if result.pvalue < 0.05:
            print(Fore.GREEN + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
        elif result.pvalue < 0.1:
            print(Fore.YELLOW + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
        else:
            print(Fore.RED + f' r={result.rvalue:.2f}, p={result.pvalue:.3f}')
        print(Style.RESET_ALL)
        #define a color based on marker
        if mu == 's': color = 'green'
        elif mu == 'o': color = 'orange'
        elif mu == '^': color = 'cyan'
        #draw
        line = f'{sub.MUSCLE.iloc[0]}_ r={result.rvalue:.2f}, p={result.pvalue:.3f}, N={len(x)}'
        axs[1, 0].scatter(x, y, marker=mu, c=color)
        axs[1, 0].plot(x, result.intercept + result.slope * x, color=color, label=line)
        for i, txt in enumerate(sub['ID']):
            axs[1, 0].annotate(chr(int(txt) + 97), (x[i]+1, y[i]))
    axs[1, 0].set_title('MUSCLE')
    axs[1, 0].set_ylabel(c)
    axs[1, 0].set_xlabel(dep)
    axs[1, 0].legend(facecolor='white', loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=1)
    # set the spacing between subplots
    plt.subplots_adjust(left=0.1,
                        bottom=0.1, 
                        right=0.9, 
                        top=0.9, 
                        wspace=0.4, 
                        hspace=0.8)
    plt.show()

#show dataframe we work with
print('Process done with following dataframe:')
reg_df
```

```python
#########################
# Stepwise regression & Multiple Linear Regression | MODELISATION
########################
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn import linear_model

"""
In this section we perform independently (meaning for all GROUP & mUSCLE subgroups) stepwise and multiple regression successively.
Stepwise regression is first performed for finding out what variables among those chosen at first ar best correlating with dependant variable (SWE).
Secondly, multiple regression is done on selected variables and coefficients are stored for further modelisation. Metrics, such as RÂ² is also saved.

The logic is to get coefficients that enable calculating a score (for instance called Charge Interne) that will predict our dependant variable (SWE).
"""
#get a working dataframe
mreg_df = df.copy()
#set colors
col = mreg_df['MUSCLE'].replace('SM', 'green').replace('ST', 'orange').replace('BF', 'cyan')

#define variables
indep = ['ACTIVATION', 'LRELATIVE', 'F_V', 'ForceRelative']
dep = 'SWE_raw'
#define number of features to retain
K = 3 #can also be an integer

#prepare a dataframe to store results
res_cols = ['Group', 'Muscle', 'n_features', 'r2', 'intercept'] + indep + ['Equation']
results = pd.DataFrame(columns=res_cols)

#work based on categorical variables
for gr, sub in mreg_df.groupby('GROUPE'):
    for mu, sub2 in sub.groupby('MUSCLE'):
        #create independent/dependant variable dataframe
        X = sub2[indep]
        y = sub2[dep]
        # Perform stepwise regression with K features to know what variables to keep
        sfs = SequentialFeatureSelector(linear_model.LinearRegression(),
                                        k_features=K,
                                        floating=True,
                                        forward=True,
                                        scoring='r2',
                                        cv=None)
        sf = sfs.fit(X, y)
        #Keep only selected variables from stepwise analysis
        X = X[list(sf.k_feature_names_)]
        #Fit a linear model using train data
        lr = LinearRegression()
        lr.fit(X, y)
        #get the predicted data from training
        y_pred = lr.predict(X)
        r2 = np.around(r2_score(y, y_pred), decimals=2)
        #Store results
        row = results.shape[0]
        results.at[row, 'Group'] = gr
        results.at[row, 'Muscle'] = mu
        results.at[row, 'n_features'] = lr.n_features_in_
        results.at[row, 'intercept'] = lr.intercept_
        results.at[row, 'r2'] = r2
        eq = str(np.around(lr.intercept_, decimals=2))
        for l, ll in enumerate(lr.feature_names_in_):
            results.at[row, ll] = lr.coef_[l]
            eq += " + " + lr.feature_names_in_[l] + "*" + str(np.around(lr.coef_[l], decimals=2))
        results.at[row, 'Equation'] = eq
        #prompt results as well
        print(f'{gr}*{mu}: {eq}', end="")
        if r2 > .4:
            print(Fore.GREEN + f'\t{r2}')
        else:
            print(Fore.RED + f'\t{r2}')
        print(Style.RESET_ALL)
    #create independent/dependant variable dataframe
    X = sub[indep]
    y = sub[dep]
    # Perform stepwise regression with K features to know what variables to keep
    sfs = SequentialFeatureSelector(linear_model.LinearRegression(),
                                    k_features=K,
                                    floating=True,
                                    forward=True,
                                    scoring='r2',
                                    cv=None)
    sf = sfs.fit(X, y)
    #Keep only selected variables from stepwise analysis
    X = X[list(sf.k_feature_names_)]
    #Fit a linear model using train data
    lr = LinearRegression()
    lr.fit(X, y)
    #get the predicted data from training
    y_pred = lr.predict(X)
    r2 = np.around(r2_score(y, y_pred), decimals=2)
    #Store results
    row = results.shape[0]
    results.at[row, 'Group'] = gr
    results.at[row, 'n_features'] = lr.n_features_in_
    results.at[row, 'intercept'] = lr.intercept_
    results.at[row, 'r2'] = r2
    eq = str(np.around(lr.intercept_, decimals=2))
    for l, ll in enumerate(lr.feature_names_in_):
        results.at[row, ll] = lr.coef_[l]
        eq += " + " + lr.feature_names_in_[l] + "*" + str(np.around(lr.coef_[l], decimals=2))
    results.at[row, 'Equation'] = eq
    #prompt results as well
    print(f'{gr:<6}: {eq}', end="")
    if r2 > .4:
        print(Fore.GREEN + f'\t{r2}')
    else:
        print(Fore.RED + f'\t{r2}')
    print(Style.RESET_ALL)
for mu, sub in mreg_df.groupby('MUSCLE'):
    #create independent/dependant variable dataframe
    X = sub[indep]
    y = sub[dep]
    # Perform stepwise regression with K features to know what variables to keep
    sfs = SequentialFeatureSelector(linear_model.LinearRegression(),
                                    k_features=K,
                                    floating=True,
                                    forward=True,
                                    scoring='r2',
                                    cv=None)
    sf = sfs.fit(X, y)
    #Keep only selected variables from stepwise analysis
    X = X[list(sf.k_feature_names_)]
    #Fit a linear model using train data
    lr = LinearRegression()
    lr.fit(X, y)
    #get the predicted data from training
    y_pred = lr.predict(X)
    r2 = np.around(r2_score(y, y_pred), decimals=2)
    #Store results
    row = results.shape[0]
    results.at[row, 'Muscle'] = mu
    results.at[row, 'n_features'] = lr.n_features_in_
    results.at[row, 'intercept'] = lr.intercept_
    results.at[row, 'r2'] = r2
    eq = str(np.around(lr.intercept_, decimals=2))
    for l, ll in enumerate(lr.feature_names_in_):
        results.at[row, ll] = lr.coef_[l]
        eq += " + " + lr.feature_names_in_[l] + "*" + str(np.around(lr.coef_[l], decimals=2))
    results.at[row, 'Equation'] = eq
    #prompt results as well
    print(f'{mu:<6}: {eq}', end="")
    if r2 > .4:
        print(Fore.GREEN + f'\t{r2}')
    else:
        print(Fore.RED + f'\t{r2}')
    print(Style.RESET_ALL)
#save results
results.to_csv(fname + '_coefficients.csv', sep=';',index=False)

"""
#Represent predictive data against actual data (from training pool) for validation
plt.scatter(y, y_pred, color=col, label=f'RÂ²={r2:.2f}')
for i, ind in enumerate(y.index):
    txt = mreg_df.loc[ind, 'ID']
    plt.annotate(chr(int(txt) + 97), (y[ind]+1, y_pred[i]))
#plt.plot(y_train, lr.intercept_ + lr.coef_ * y_train, label=line)
plt.xlabel('Experimental Data')
plt.ylabel('Predicted Data')
plt.legend()
plt.show()
"""
```
